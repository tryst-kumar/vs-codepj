```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
```

```python
train = pd.read_csv('train.csv')
```

```python
train.head()
```

```python
le = LabelEncoder()
train['Sex'] = le.fit_transform(train['Sex'])
```

```python
train.head()
```

```python
X = train.drop(columns=['id', 'Calories'])
y = train['Calories']
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
val_preds = lr_model.predict(X_val)
rmse = mean_squared_error(y_val, val_preds)
r2 = r2_score(y_val, val_preds)
print(f"Linear Regression RMSE: {rmse:.2f}")
print(f"Linear Regression R2 Score: {r2:.4f}")
```

linear legression prediction is good but steel i would  like to refine it

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtree_model = DecisionTreeRegressor()
dtree_model.fit(X_train, y_train)
```

```python
val_predicted = dtree_model.predict(X_val)
```

```python
rmse2 = mean_squared_error(y_val, val_predicted)
r2_2 = r2_score(y_val, val_predicted)

print(f"Decision tree RMSE {rmse2:.2f}")
print(f"Decision Tree r2 score {r2_2:.4f}")
```

here the accuracy is much better compared to linear regression

now lets move to RandomforestRegressior

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr_model = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
rfr_model.fit(X_train, y_train)
```

```python
rf_val_predicts = rfr_model.predict(X_val)
```

```python
rmse3 = mean_squared_error(y_val, rf_val_predicts)
r2_3 = r2_score(y_val, rf_val_predicts)

print(f"Mean Squared Error {rmse3:.2f}")
print(f"Root mean square: {r2_3:.4f}")
```

The accuracy is very much decent
